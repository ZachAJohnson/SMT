# YVM and IYVM Validation

[1] "Viscosity estimates of liquid metals and warm dense matter using the Yukawa reference system"  [doi.org/10.1016/j.hedp.2007.11.001](https://doi.org/10.1016/j.hedp.2007.11.001)

[2]  "Comparison of Transport Models in Dense Plasmas" 

[3] "Determination of the shear viscosity of the one-component plasma" [doi/10.1103/PhysRevE.90.033105](https://journals.aps.org/pre/pdf/10.1103/PhysRevE.90.033105)

```python
import numpy as np
import matplotlib.pyplot as plt

from pandas import read_csv

import SMT.core.transport as SMT
from SMT.core.physical_constants import *
from SMT.core.physical_parameters import rs_from_n
```

### Import Daligault data from [3]

```python
Dal_κ0_data = read_csv("Daligault_viscosity_OCP_κ=0.txt", delim_whitespace=True, header=2)
Dal_κ2_data = read_csv("Daligault_viscosity_OCP_κ=2.txt", delim_whitespace=True, header=2)
```

## Choose κ and Generate (I)YVM Opinion

### κ=0

```python
κ_test = 0
data_to_compare = Dal_κ0_data
Γ_array = data_to_compare['Γ']
η_reference =  data_to_compare['η/(mna^2gωp)']
dense_Γ_array = np.geomspace(5e-3, 1e3, num=1000)
```

```python
Γ_melt = lambda κ: 171.8 + 82.8*(np.exp(0.565*κ**1.38) - 1) 

# Actual YVM formula
η_YVM_func = lambda Γ, κ: np.exp(-0.2*κ**1.62)*(0.0051*Γ_melt(κ)/Γ + 0.374*Γ/Γ_melt(κ) + 0.022)
# η_YVM_array = 
@np.vectorize

def η_IYVM_func(Γ,κ):
    A = 1.45e-4 - 1.04e-4*κ + 3.69e-5*κ**2
    B = 0.3 + 0.86*κ-0.69*κ**2 + 0.138*κ**3
    C = 0.015 + 0.048*κ*0.754
    a = 1.78 + 0.13*κ - 0.062*κ**2
    b = 1.63 - 0.325*κ + 0.24*κ**2

    # Actual IYVM formula
    return np.exp(-0.2*κ**1.62) * ( A*(Γ_melt(κ)/Γ)**a + B*(Γ/Γ_melt(κ))**b + C )


# Daligault Fit to OCP
@np.vectorize
def η_Daligault_func(Γ):
    a, b, a1, a2, a3 =  0.794811, 0.862151, 0.0425698, 0.00205782, 7.03658e-5
    b1, b2, b3, b4 =  0.0429942, -0.000270798, 3.25441e-6, -1.15019e-8

    pre = a/(Γ**2.5*np.log(1 + b/Γ**1.5) ) 
    num = 1 + a1*Γ**1 + a2*Γ**2 + a3*Γ**3
    denom = 1 + b1*Γ*1 + b2*Γ**2 + b3*Γ**3 + b4*Γ**4
    return pre * num / denom

# η_Daligault_array =  η_Daligault_func(dense_Γ_array)
# Landau-Spitzer
η_LS_func = lambda Γ: 0.794811* Γ**-2.5 * 1/np.log( 1.493 * 1/np.sqrt(3) * Γ**-1.5  )
η_LS_Dailgault_func = lambda Γ: 0.794811* Γ**-2.5 * 1/np.log(  1 + 1.493 * 1/np.sqrt(3) * Γ**-1.5  )
```

#### Plot

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8, 9), sharex=True, gridspec_kw={'height_ratios': [3, 1]})

# Plotting the main comparison
ax1.plot(dense_Γ_array, η_YVM_func(dense_Γ_array, 0 ), label="YVM")
ax1.plot(dense_Γ_array, η_IYVM_func(dense_Γ_array, 0 ), label="IYVM")
ax1.plot(dense_Γ_array, η_Daligault_func(dense_Γ_array), label="Daligault 2014 Pade Fit ")

ax1.plot(Γ_array, η_reference, 'ko', label='Daligault 2014')
ax1.plot(dense_Γ_array, η_LS_func(dense_Γ_array), 'r--', label="Landau-Spitzer")
ax1.plot(dense_Γ_array, η_LS_Dailgault_func(dense_Γ_array), 'b--', label="Landau-Spitzer + Daligault 2014")

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.legend()
ax1.set_ylabel(r"$\eta^*$", fontsize=14)
ax1.set_title(r"(I)YVM Comparison to Daligault 2014: $\kappa=0$", fontsize=12)
ax1.tick_params(labelsize=12)
ax1.set_ylim(3e-2, 2e4)
ax1.set_xlim(np.min(dense_Γ_array), np.max(dense_Γ_array))

# Calculate and plot the relative error
relative_error_YVM = (η_YVM_func(Γ_array, 0) - η_reference) / η_reference
relative_error_IYVM = (η_IYVM_func(Γ_array, 0) - η_reference) / η_reference
relative_error_Daligault = (η_Daligault_func(Γ_array) - η_reference) / η_reference

ax2.plot(Γ_array, relative_error_YVM, label="Relative Error YVM")
ax2.plot(Γ_array, relative_error_IYVM, label="Relative Error IYVM")
ax2.plot(Γ_array, relative_error_Daligault, label="Relative Error Daligault")

ax2.axhline(0, color='gray', linestyle='--')  # Add a horizontal line at zero error
ax2.set_xscale('log')
ax2.set_xlabel(r"$\Gamma$", fontsize=14)
ax2.set_ylabel("Relative Error", fontsize=14)
ax2.tick_params(labelsize=12)
ax2.set_ylim(-1,1)
ax2.set_xlim(np.min(dense_Γ_array), np.max(dense_Γ_array))

plt.tight_layout()
# plt.show()
plt.savefig("YVM_to_Daligault_κ=0.png", dpi=200)
```
